```python
# Define a custom class demonstrating object-oriented programming concepts
# This class demonstrates type hints, private attributes, and error handling
class MyClass:
    def __init__(self, value: int | str):
        # Private attribute (name mangling with double underscore)
        self.__private_value = 42
        
        # Try to convert input to integer, otherwise store as string
        try:
            self.value = int(value)
        except ValueError:
            print("Value is not an integer, storing as string.")
            self.value = value

    def display(self):
        """Display the current value of the object."""
        print(f"Value: {self.value}")
        
    def __private_method(self):
        """This method is private and cannot be called directly from outside."""
        print("This is a private method.")
        pass
```

```python
# Instantiate MyClass with different value types
# c will store an integer, b will store a string
c = MyClass(10)
b = MyClass("Test")
b.display()
c.display()
# Note: Uncommenting the lines below would fail because __private_method and __private_value are private
# c.__private_method()
# c.__private_value
```

```python
# Classic Python interview gotcha - mutable default arguments
# This demonstrates a common pitfall: using mutable objects as default parameter values
# The default list is created ONCE when the function is defined, not on each call!
def append_to_list(value, target_list=[]):
    print(f"List object id: {id(target_list)}")  # Same id every time!
    target_list.append(value)
    return target_list

# Each call reuses the same list object
result1 = append_to_list(1)
print(f"result1: {result1}")
result2 = append_to_list(2)
result3 = append_to_list(3)

# Surprising behavior: all results point to the same list
print(f"result1: {result1}")  # [1, 2, 3] - unexpected!
print(f"result2: {result2}")  # [1, 2, 3] - same list
print(f"result3: {result3}")  # [1, 2, 3] - same list

# The fix - use None as default
# def append_to_list_fixed(value, target_list=None):
#     if target_list is None:
#         target_list = []
#     target_list.append(value)
#     return target_list
```

```python
# Demonstrate Python's reference semantics and copy behavior
# This shows the difference between assignment and copying

# Create a list
a = [1, 2, 3]
# Assign b to reference the same list (not a copy)
b = a
b.append(4)
print("After b.append(4):")
print(f"a: {a}")  # Also changed because a and b reference the same list
print(f"b: {b}")
print(f"a is b: {a is b}")  # True - same object in memory

# Create a copy using .copy()
c = a.copy()  # Shallow copy
c.append(5)
print("\nAfter c.append(5):")
print(f"a: {a}")  # Unchanged
print(f"c: {c}")  # Different list
print(f"a is c: {a is c}")  # False - different objects
```

## Working with Neural Data (JSON Example)

The following section demonstrates how to load and process structured data from a JSON file containing neural recordings. This is a practical example of data loading, transformation, and basic visualization.

```python
# Load neural data from a JSON file
# JSON is a common format for structured data interchange
import json

# Read the JSON file containing neural recordings
with open('/home/miksolo/git/CerebroV1/cerebro_embedded_esp32-s3_v1/scripts/neural_data/neural_data_20251009_222632.json', 'r') as f:
    neural_data = json.load(f)

# Display a summary of the loaded data (first few characters to avoid overwhelming output)
print(f"Loaded neural data with {len(neural_data)} top-level keys")
print(f"Keys: {list(neural_data.keys())}")
```

```python
# Examine the structure of the neural data
# This shows the top-level keys in the dictionary
print("Top-level keys in neural_data:")
neural_data.keys()
```

```python
# Inspect the session information metadata
print("Session information:")
neural_data['session_info']
```

```python
# Examine a specific cycle from the neural data
# Shows the structure of individual measurement records
print("Sample cycle data (message 10, cycle 38):")
neural_data['messages'][10]['data']['cycles'][38]
```

```python
# Transform nested JSON data into a flat pandas DataFrame
# This is a common pattern: extract nested data and organize it for analysis
all_cycles = []

# Iterate through all messages and extract cycle data
for message in neural_data['messages']:
    if 'data' in message and 'cycles' in message['data']:
        for cycle in message['data']['cycles']:
            all_cycles.append(cycle)

# Convert the list of dictionaries to a pandas DataFrame
# This allows us to use powerful data manipulation and analysis tools
df_neural = pandas.DataFrame(all_cycles)

print(f"Created DataFrame with {len(df_neural)} cycles and {len(df_neural.columns)} columns")
df_neural
```

```python
# Visualize the distribution of voltage measurements
# This creates a histogram showing how voltage values are distributed
import matplotlib.pyplot as plt

# Create histogram
plt.hist(df_neural['v'], bins=30)
plt.xlabel('Voltage (v)')
plt.ylabel('Frequency')
plt.title('Distribution of Voltage (v) in Neural Data')
plt.show()

print(f"Min voltage: {df_neural['v'].min():.4f} V")
print(f"Max voltage: {df_neural['v'].max():.4f} V")
print(f"Mean voltage: {df_neural['v'].mean():.4f} V")
```
